```python
def plot_coronagraph_spectrum(wl, ofrat, sig, 
                              truth=None, 
                              xlim=None, ylim=None, 
                              title="", 
                              save=False, tag=""):
    
    fig = plt.figure(figsize=(12,10))
    gs = gridspec.GridSpec(1,1) 
    ax = plt.subplot(gs[0])

    if truth is not None:
        ax.plot(wl, truth*1e9, lw=2.0, color="purple", alpha=0.7, ls="steps-mid")
    ax.errorbar(wl, ofrat*1e9, yerr=sig*1e9, fmt='o', color='k', ms=5.0)

    ax.set_ylabel(r"F$_p$/F$_s$ ($\times 10^9$)")
    ax.set_xlabel("Wavelength [$\mu$m]")
    ax.set_title(title)

    if ylim is not None: ax.set_ylim(ylim)
    if xlim is not None: ax.set_xlim(xlim)
        
    if save:
        fig.savefig(title+tag+".pdf")
```

```python
plot_coronagraph_spectrum(lam, spec, sig)
```
